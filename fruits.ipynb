```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
fruits = pd.DataFrame({
    'A': np.random.choice(np.array((1,2,3)), 15),
    'B': np.random.choice(['apple', 'mango', 'banana', 'kiwi', 'pineapple', 'peach'], 15, True),
    'C': np.round(50 + 5*np.random.randn(15), 2)
})
```

```python
fruits
```

```python
fruits.head()
```

```python
sns.countplot(y = "B", data = fruits, color = "black")
```

```python
sns.pairplot(fruits)
```
